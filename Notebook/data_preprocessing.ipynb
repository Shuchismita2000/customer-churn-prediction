### Encoding

```python
categorical_columns = ['City', 'Zip Code', 'Gender', 'Senior Citizen', 'Partner', 'Dependents', 'Phone Service', 'Multiple Lines',
                        'Internet Service', 'Online Security', 'Online Backup', 'Device Protection', 'Tech Support',
                        'Streaming TV', 'Streaming Movies', 'Contract', 'Paperless Billing', 'Payment Method', 'Churn Reason', 'Churn Label']
```

```python
# Step 2: Use label encoding to encode the categorical columns
from sklearn.preprocessing import LabelEncoder

label_encoders = {}
for col in categorical_columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
    #dump(encoder{col}, 'pipeline/encoder{col}.joblib')

```

```python
# Fit the LabelEncoders (assuming X_train is your training data with these columns)
for column, le in label_encoders.items():  
    dump(le, f'pipeline/{column}_label_encoder.joblib')
```

### Scale Numerical Features

```python
df.columns
```

```python
from sklearn.preprocessing import StandardScaler

numerical_features = ['Tenure Months', 'Monthly Charges', 'Total Charges']
scaler = StandardScaler()
df[numerical_features] = scaler.fit_transform(df[numerical_features])

dump(scaler, 'pipeline/scaler.joblib')
```

### Feature Selection

```python
df.drop(['Unnamed: 0','CustomerID','Count', 'Churn Reason', 'Churn Value', 'Churn Score'], axis=1,inplace=True)
```

```python
X = df.drop('Churn Label', axis=1)
y = df['Churn Label']
```

Correlation Analysis

```python
# Calculate the correlation matrix
corr_matrix = X.corr().abs()
corr_matrix
```

```python
import numpy as np

# Select the upper triangle of the correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
```

```python
# Find features with correlation greater than 0.8
to_drop = [column for column in upper.columns if any(upper[column] > 0.8)]
```

```python
# Drop highly correlated features
X = X.drop(columns=to_drop)
```

```python
X 
```

Recursive Feature Elimination (RFE)

Use RFE with a chosen model to recursively remove features with the least importance.

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier

# Instantiate a RandomForestClassifier
model = RandomForestClassifier()

# Instantiate RFE with the model
rfe = RFE(estimator=model, n_features_to_select=10) # choose number of features

# Fit RFE
rfe.fit(X, y)

# Select features
X_rfe = X.loc[:, rfe.support_]
```
